# Time Series Analysis

```python
from statsmodels import tsa as ts
import pandas as pd
import numpy as np
from ggplot import *
# Enable inline plotting
%matplotlib inline
```

```python
d = np.random.normal(size=300)
#D = ts.filters.bkfilter(d, K=8)

D = ts.ar_model.AR(d)
D.fit()
```

```python
qplot(y=d, x=xrange(len(d)))+geom_line()
```

```python
qplot(y=D, x=xrange(len(D)))+geom_line()
```
